```python
#1
# 定义一个函数 Print_values，接收三个参数 a, b, c
def Print_values(a, b, c):
    # 当 a 大于 b 时，进入循环
    while a > b:
        # 如果 b 大于 c，执行以下语句
        if b > c:
            # 输出表达式 a + b - 10 * c 的结果
            print(a + b - 10 * c)
        else:
            # 如果 a 大于 c，执行以下语句
            if a > c:
                # 输出表达式 a + c - 10 * b 的结果
                print(a + c - 10 * b)
            else:
                # 否则，输出表达式 c + a - 10 * b 的结果
                print(c + a - 10 * b)
    # 如果 while 条件不满足（即 a <= b），执行 else 块
    else:
        # 如果 b 大于 c，输出字符串 'END'
        if b > c:
            print('END')
        else:
            # 否则，输出表达式 c + b - 10 * a 的结果
            print(c + b - 10 * a)

# 调用函数，传入参数 a=5, b=15, c=10
Print_values(5, 15, 10)
```

```python
#2
# 导入 math 模块，用于使用 math.ceil 函数进行向上取整
import math
# 定义主函数 Cal，接收一个整数列表 N
def Cal(N):
    # 定义内部函数 F(x)，用于计算连续取整函数 F(x) = F(ceil(x/3)) + 2x，且 F(1) = 1
    def F(x):
        # 基础情况：F(1) = 1
        if x == 1:             
            return 1
        # 递归调用：对 x 进行向上取整除以 3，然后加上 2x
        return F(math.ceil(x / 3)) + 2 * x
    # 输入验证：确保列表中的每个元素都是正整数
    for x in N:
        # 如果不是整数或小于等于 0
        if int(x) != x or x <= 0:
            # 提示错误信息并返回 None
            return print("Given a list with N positive integers please")  
    # 如果所有输入合法，返回每个元素对应的 F(x) 结果列表
    return [F(x) for x in N]
print(Cal([1, 2, 3]))
```

```python
#4.2
import random
# 导入绘图库，用于绘制趋势图
import matplotlib.pyplot as plt
# 定义函数：生成一个长度为 N 的整数数组，元素在 1 到 10 之间随机选取
def Random_integer(N):
    return [random.randint(1, 10) for x in range(N)]
# 定义函数：计算所有非空子集的平均值之和
def Sum_averages(arr):
    # 初始化子集列表，开始时只有一个空子集
    subsets = [[]]  
    # 初始化总和为 0
    total = 0       
    # 遍历数组中的每个元素
    for num in arr:
        # 用于存储新生成的子集
        new_subsets = []  
        for subset in subsets:
            # 将当前元素加入已有子集，形成新子集
            new_subset = subset + [num] 
            # 添加到新子集列表中
            new_subsets.append(new_subset)  
            # 计算新子集的平均值
            avg = sum(new_subset) / len(new_subset)  
            # 累加到总和中
            total += avg  
        # 将新子集加入原子集列表，继续扩展
        subsets.extend(new_subsets)  
    # 返回所有非空子集平均值的总和
    return total  
arr = Random_integer(10)
result = Sum_averages(arr)
print("随机数组：", arr)
print("所有非空子集平均值之和：", result)
```

```python
#5.2
import random
import numpy as np
# 创建一个 N 行 M 列的矩阵，左上角和右下角固定为 1，其余位置随机为 0 或 1
def Create_matrix(N, M):
    # 使用列表推导式生成随机矩阵，然后转换为 numpy 数组
    matrix = np.array([[random.randint(0, 1) for x in range(M)] for y in range(N)])
    # 设置左上角为 1（起点）
    matrix[0][0] = 1   
    # 设置右下角为 1（终点）
    matrix[N-1][M-1] = 1   
    # 返回生成的矩阵
    return matrix 
# 计算从左上角到右下角的所有合法路径数（只能向右或向下走，且只能走在值为 1 的格子上）
def Count_path(matrix):
    # 获取矩阵的行数和列数
    N, M = len(matrix), len(matrix[0])  
    # 如果起点或终点是障碍（值为 0），直接返回 0 条路径
    if matrix[0][0] == 0 or matrix[N-1][M-1] == 0:
        return 0
    # 初始化路径列表，起点为 (0, 0)
    paths = [[(0, 0)]] 
    # 初始化合法路径计数器
    count = 0                   
    # 使用 while 循环不断扩展路径，直到没有路径可扩展
    while paths:
        # 存储下一轮扩展后的路径
        new_paths = []           
        for path in paths:
            # 获取当前路径的最后一个位置
            i, j = path[-1]      
            # 如果到达终点，计数加一
            if (i, j) == (N - 1, M - 1):
                count += 1
                continue
            # 尝试向右扩展路径
            if j + 1 < M and matrix[i][j + 1] == 1:
                new_paths.append(path + [(i, j + 1)])
            # 尝试向下扩展路径
            if i + 1 < N and matrix[i + 1][j] == 1:
                new_paths.append(path + [(i + 1, j)])
        # 更新当前路径列表为扩展后的路径
        paths = new_paths        
    # 返回合法路径总数
    return count                 
print("矩阵：")
print(matrix)
print("合法路径数：", Count_path(matrix))
```

```python
#5.3
import random
import numpy as np
# 创建一个 N 行 M 列的矩阵，左上角和右下角固定为 1，其余位置随机为 0 或 1
def Create_matrix(N, M):
    # 使用列表推导式生成随机矩阵，然后转换为 numpy 数组
    matrix = np.array([[random.randint(0, 1) for x in range(M)] for y in range(N)])
    # 设置左上角为 1（起点）
    matrix[0][0] = 1   
    # 设置右下角为 1（终点）
    matrix[N-1][M-1] = 1   
    # 返回生成的矩阵
    return matrix 
# 计算从左上角到右下角的所有合法路径数（只能向右或向下走，且只能走在值为 1 的格子上）
def Count_path(matrix):
    # 获取矩阵的行数和列数
    N, M = len(matrix), len(matrix[0])  
    # 如果起点或终点是障碍（值为 0），直接返回 0 条路径
    if matrix[0][0] == 0 or matrix[N-1][M-1] == 0:
        return 0
    # 初始化路径列表，起点为 (0, 0)
    paths = [[(0, 0)]] 
    # 初始化合法路径计数器
    count = 0                   
    # 使用 while 循环不断扩展路径，直到没有路径可扩展
    while paths:
        # 存储下一轮扩展后的路径
        new_paths = []           
        for path in paths:
            # 获取当前路径的最后一个位置
            i, j = path[-1]      
            # 如果到达终点，计数加一
            if (i, j) == (N - 1, M - 1):
                count += 1
                continue
            # 尝试向右扩展路径
            if j + 1 < M and matrix[i][j + 1] == 1:
                new_paths.append(path + [(i, j + 1)])
            # 尝试向下扩展路径
            if i + 1 < N and matrix[i + 1][j] == 1:
                new_paths.append(path + [(i + 1, j)])
        # 更新当前路径列表为扩展后的路径
        paths = new_paths        
    # 返回合法路径总数
    return count   
# 设置矩阵大小为 10 行 8 列
N, M = 10, 8
# 设置重复运行次数为 1000
runs = 1000   
# 初始化总路径数
total_paths = 0                 
# 重复运行 1000 次，每次生成新矩阵并计算路径数
for _ in range(runs):
    # 生成随机矩阵
    matrix = Create_matrix(N, M)  
    # 累加路径数
    total_paths += Count_path(matrix)   
# 计算平均路径数
average_paths = total_paths / runs
# 打印结果
print("1000 次运行的平均路径数：", average_paths)
```
